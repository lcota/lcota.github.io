```r
library(data.table)
library(quantmod)
library(xts)
```

symbols from yahoo finance screener:  
https://finance.yahoo.com/screener/0ea79e7a-0ac8-4d4c-af49-d503ec7090ec

```r
syms = read.csv("data/syms.csv", header=T)$Symbol
getSymbols(syms)

dt <- data.table()
for (i in 1:length(syms)) {
    sym = syms[i]
    data = get(sym)
    data = as.data.table(data)
    colnames(data) <- c('date', 'open', 'high', 'low', 'close', 'volume', 'adjusted')
    data[, sym:=sym]
    if ( is.null(dt)){
        dt <- data
    }
    else {
        dt <- rbindlist(list(dt, data))
    }
}

dt[, mid:= (high + low)*.5]

fwrite(dt, file="data/raw/etfs.csv", append=F, sep=',')
print("DONE")
```
